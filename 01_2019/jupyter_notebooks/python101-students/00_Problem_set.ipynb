## 1. 이상한 덧셈
> 숫자로 구성된 리스트에서 양의 정수의 합을 구하는 함수, `positive_sum`을 만들어보아요

예시)
```python
positive_sum([1, -10, 2]) # 3
positive_sum([-1, -2, -3, -4]) # 0
```

````python
def positive_sum(sum_list):
    sum_num = 0
    for num in sum_list:
        if num > 0:
            sum_num += num
    return sum_num

positive_sum([-1, -2, -3, -4])


# 선생님 답
def pos_sum(numbers):
    return sum(x for x in numbers if x > 0)

print(pos_sum([1, 2, 3, -6]))
````

### 3. Collatz
> Collatz 추측 : 어떤 자연수 n이든지, 다음과 같은 작업을 반복하면 1로 만들 수 있다.
1. n이 짝수라면 2로 나눈다.
2. n이 홀수라면 3을 곱하고 1을 더한다.
3. 결과로 나온 수에 1/2의 작업을 1이 될 때까지 반복한다.

> 예를 들어 n이 6이면, 6 => 3 => 10 => 5 => 16 => 8 => 4 => 2 => 1이 되며 8번(=>의 갯수)만에 1이 됩니다. 자연수 n이 들어왔을 때, 몇 번의 작업만에 1이 되는지를 `return`하는 함수 `collatz()`를 완성하세요.

> 단, 500번을 넘어가도 1이 되지 않는다면, -1을 `return`하세요.

예시)
```python
collatz(6) # 8
collatz(16) # 4
collatz(6263313) # -1
```

````python
def collatz(n):
    count = 0
    
    while n != 1:
        if count > 500:
            return -1

        if n % 2 == 0:
            n /= 2
            count += 1
        else:
            n = n * 3 + 1
            count += 1
            
    return count

print(collatz(6))
print(collatz(16))
print(collatz(626331))


def collatz1(n, count=0):
    
    if n == 1:
        return count
    
    if count > 500:
        return -1
    
    if n % 2 == 0:
        count += 1
        return collatz1(n//2, count)
    else:
        count += 1
        return collatz1(n*3+1, count)

        
print(collatz1(6))
print(collatz1(16))
print(collatz1(626331))
````

### 5. * RGB 삼각형

| BG   | RG   | BR   | RR   | GG   | BB   |
| ---- | ---- | ---- | ---- | ---- | ---- |
| R    | B    | G    | R    | G    | B    |

> 입력으로 `"R", "G", "B"`가 섞여 있는 문자열이 들어온다. 문자열은 다음과 같이 합쳐진다. 해당 문자열을 처리하여 마지막 색깔만 `return`하는 `triangle()`을 작성하세요.
```
R R G B R G B B
 R B R G B R B
  G G B R G G
   G R G B G
    B B R R
     B G R
      R B
       G
```
예시)
```python
triangle("RRGBRGBB") # G
triangle("GB") # R
triangle("B") # B
triangle("RGBG") # "G"
```

````python
def triangle(kwarg, kwarg2=""):

    if len(kwarg) == 1 and kwarg2 == "":
        return kwarg

    if len(kwarg) == 1:
        return triangle(kwarg2)

    if kwarg[:2] in ("RR", "GB", "BG"):
        kwarg2 += "R"
        kwarg = kwarg[1:]
        return triangle(kwarg, kwarg2)
    
    elif kwarg[:2] in ("GG", "RB", "BR"):
        kwarg2 += "G"
        kwarg = kwarg[1:]
        return triangle(kwarg, kwarg2)
    
    else:
        kwarg2 += "B"
        kwarg = kwarg[1:]
        return triangle(kwarg, kwarg2)
    
    
print(triangle("RRGBRGBB"))
print(triangle("GB"))
print(triangle("B"))
print(triangle("RGBG"))


def triangle1(kwarg, kwarg2=""):

    count = 0
    if len(kwarg) == 1:
        return kwarg

    for i in kwarg:
        if count + 1 != len(kwarg):
            if i == kwarg[count + 1]:
                kwarg2 += i
                count += 1
            elif i + kwarg[count + 1] in ("RR", "GB", "BG"):
                kwarg2 += "R"
                count += 1
            elif i + kwarg[count + 1] in ("GG", "RB", "BR"):
                kwarg2 += "G"
                count += 1
            else:
                kwarg2 += "B"
                count += 1
        else:
            return triangle1(kwarg2)

    
print(triangle1("RRGBRGBB"))
print(triangle1("GB"))
print(triangle1("B"))
print(triangle1("RGBG"))
````

````python
def find_odd(odds):
    
    set_odds = set(odds)
    
    for i in odds:
        set_odds.add(i)
        
    for i in set_odds:
        if odds.count(i) % 2:
            return i

# 선생님 답(위랑 방식은 같지만 줄이는 방법)
def find_odd1(numbers):
    for i in list(set(numbers)):
        if numbers.count(i) % 2:
            return i

def find_odd2(numbers):
    result = 0
    for n in numbers:
        result = result ^ n
    return result

# xor : 같으면 0 다르면 1

from operater import xor
def find_odd3(numbers):
    return reduce(xor, numbers)


print(find_odd2([1, 1, 2, 2, 3, 3, 3]))
print(find_odd2([2, 1, 2]))
print(find_odd2([1, 2, 2, 3, 2, 2, 1]))
````

````python
import random

def guess_up_down(s=1, e=100, count=0):

    if count == 0:
        print("숫자를 생각하세요...")
        
    N = random.choice(range(s, e + 1))
    print(f"{N}이 맞나요?")
    
    answer = input("작으면 -1, 맞으면 0, 크면 1 : ")
    
    if answer == "-1":
        count += 1
        return guess_up_down(s, N - 1, count)
    elif answer == "1":
        count += 1
        return guess_up_down(N + 1, e, count)
    else:
        count += 1
        print(f"{count}번 만에 맞혔습니다.")
        

guess_up_down()
````

````python
# 선생님 답
def is_answer(answer, letters):
    answer = set(answer)
    count = 0
    for char in set(letters):
        if char in answer:
            count += 1
    return len(answer) == count
    
````

````python
def status(answer, letters):
    
    for char in answer:
        if char not in letters:
            answer = answer.replace(char, "*")
    
    return answer

status("apple", ["a"])
````
