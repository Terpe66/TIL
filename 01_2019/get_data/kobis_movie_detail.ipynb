```python
import os
import requests

week = ["181111", "181118", "181125", "181202", "181209", "181216", "181223", "181230", "190106", "190113"]
key = os.getenv("KOBIS_KEY")
url = f"http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchWeeklyBoxOfficeList.json?key={key}&weekGb=0&targetDt=20"

```

```python
kobis_data = []
for wk in week:
    kobis_data.append(requests.get(url + wk).json()["boxOfficeResult"])
```

```python
kobis_data
```

```python
kobis_list = []
for kobis_d in kobis_data:
    for index, kobis in enumerate(kobis_d["weeklyBoxOfficeList"]):
        add = [kobis["movieCd"], kobis["movieNm"], kobis["audiAcc"], kobis_d["showRange"][9:-1]]
        if kobis_list == []:
            kobis_list.append(add)
        else:
            for idx in range(len(kobis_list)):
                if add[0] in kobis_list[idx]:
                    kobis_list[idx] = add
                    break
            else:
                kobis_list.append(add)
                
print(kobis_list)
print(len(kobis_list))
```

```python
import csv

f_w = open("boxoffice.csv", "a+", encoding="utf-8", newline="")
writer = csv.writer(f_w)

for i in kobis_list:
    writer.writerow(i)

f_w.close()
```

- 영화 대표 코드 : movieCd

- 영화명 : movieNm

- 해당일 누적관객수 : audiAcc
    - 중복되면 최신 정보를 반영해야 한다.

- 해당일 : showRange
    - showRange는 "20181105~20181111"식이라서 맨 마지막 8자리를 구해야 할 듯.

```python
key = os.getenv("KOBIS_KEY")

f_r = open("boxoffice.csv", "r", encoding="utf-8", newline="")
reader = csv.reader(f_r)

movie_info = []
for line in reader:
    url = f"http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key={key}&movieCd={line[0]}"
    movie_info.append(requests.get(url).json())
    
    
f_r.close()

print(movie_info)

```

- 영화명(원문) : movieNmOg

- 개봉연도 : openDt

- 상영시간 : showTm

- 장르 : genreNm

- 감독명 : peopleNm

- 관람등급 : watchGradeNm
